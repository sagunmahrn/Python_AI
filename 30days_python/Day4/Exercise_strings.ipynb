```python
# Create an acronym or an abbreviation for the name 'Python For Everyone'.
phrase = "Python For Everyone"

# Split into words
words = phrase.split()

# Build acronym: "Python" → "Py", "For" → "4", "Everyone" → "E"
acronym = words[0][:2] + "4" + words[2][0].upper()

print("Creative Acronym:", acronym)
```

```python
# Create an acronym or an abbreviation for the name 'Coding For All'.
phrase2 = "Coding For All"
# Split into words
words2 = phrase2.split()
# Build acronym: "Coding" → "C", "For" → "F", "All" → "A"
acronym2 = words2[0][0].upper() + words2[1][0].upper() + words2[2][0].upper()
print("Acronym:", acronym2)
```
