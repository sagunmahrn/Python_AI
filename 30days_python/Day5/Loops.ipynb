```python
# Iterate 0 to 10 using for loop, do the same using while loop.

i=0
while i <= 10:
    print(i)
    i += 1 
```

```python
# Iterate 10 to 0 using for loop, do the same using while loop.
i=10
while i >= 0:
    print(i)
    i -= 1
```

```python
# write a program to print all even numbers from 0 to 50 using while loop.
# (Hint: use % operator to check for even numbers)
# Example output: 0,2,4,6,...,48,50

num = 1
while num <= 50:
    if num % 2 == 0:
        print(num)
    num += 1

```

```python
# write a program that prints the sum of first n natural numbers.
# for example, if n=5, the output should be 15 (1+2+3+4+5)
number = int(input("Enter a number: "))
sum = 0
i=1
while(number>=i):
    sum=sum+number
    number = number-1
print("Sum:",sum)
print("Number:",number)
```

```python
# Write a program to print this pattern using a while loop.
""" *
    **
    ***
    ****
    *****
"""
n = 1
while n <= 5:
    print('*' * n)
    n += 1
```

```python
# Write a loop that makes seven calls to print(), so we get on the output the following triangle:
n=1
while n <= 7:
    print('#' * n)
    n += 1
```

```python
# Saumya want to print her name 5 times,but each time with a number in front of it.Write a program to help her do that using while loop.
count = 1
while count <= 5:
    print(f"{count}. Saumya Singh")
    count += 1
```

```python
# Use nested loops to create the following:

"""
# # # # # #
# # # # # #
# # # # # #
# # # # # #
# # # # # #
# # # # # #
"""
count = 1
while count <= 6:
    print('# ' * 6)
    count += 1
```
