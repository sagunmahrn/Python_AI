```python
# write a program that prints the sum of first n natural numbers.
# for example, if n=5, the output should be 15 (1+2+3+4+5)
number = int(input("Enter a number: "))
sum = 0
i=1
while(number>=i):
    sum=sum+number
    number = number-1
print("Sum:",sum)
print("Number:",number)
```

```python
#Use for loop to iterate from 0 to 100 and print the sum of all evens and the sum of all odds.
sum_even = 0
sum_odd = 0
for i in range(0, 101):
    if i % 2 == 0:
        sum_even += i
    else:
        sum_odd += i
print("Sum of all even numbers from 0 to 100 is:", sum_even)
print("Sum of all odd numbers from 0 to 100 is:", sum_odd)
```

```python
# This is a fruit list, ['banana', 'orange', 'mango', 'lemon'] reverse the order using loop.
fruits = ['banana', 'orange', 'mango', 'lemon']
reversed_fruits = []
for fruit in fruits:
    reversed_fruits.insert(0, fruit)
print(reversed_fruits)
```

```python
# counter timer
import time

count = int(input("Enter a counter number: "))

print("Counting down:")
for i in range(count, 0, -1):
    print(i)
    time.sleep(1)
print("Countdown finished!")
```

```python
# Guess the number game
number_to_guess =int(input("Enter a number to guess between 1 and 10: "))
while True:
    user_guess = int(input("Your guess: "))
    if user_guess < number_to_guess:
        print("Too low! Try again.")
    elif user_guess > number_to_guess:
        print("Too high! Try again.")
    else:
        print("Congratulations! You guessed the number.")
        break
```
